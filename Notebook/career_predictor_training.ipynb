```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import joblib

# STEP 1: Load your dataset
df = pd.read_csv("../data/student_career_dataset.csv")

# STEP 2: Fill missing values with 0 (recommended by your lecturer for real gaps)
df.fillna(0, inplace=True)

# STEP 3: Assign career label based on highest score
def assign_label(row):
    scores = {
        "DB": row["avg_db"],
        "AI": row["avg_ai"],
        "WEB": row["avg_web"],
        "SYS": row["avg_sys"],
        "CS": row["avg_cs"]
    }
    return max(scores, key=scores.get)

df["career_label"] = df.apply(assign_label, axis=1)

df
```

```python
# STEP 3: Assign career label based on highest score
def assign_label(row):
    scores = {
        "DB": row["avg_db"],
        "AI": row["avg_ai"],
        "WEB": row["avg_web"],
        "SYS": row["avg_sys"],
        "CS": row["avg_cs"]
    }
    return max(scores, key=scores.get)

df["career_label"] = df.apply(assign_label, axis=1)
```

```python

# STEP 4: Prepare features and labels
X = df[["avg_db", "avg_ai", "avg_web", "avg_sys", "avg_cs"]]
y = df["career_label"]

# STEP 5: Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# STEP 6: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python

# STEP 7: Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# STEP 8: Evaluate model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
joblib.dump(model, "../final_app/models/career_predictor_model.pkl")
joblib.dump(le, "../final_app/models/label_encoder.pkl")
```
